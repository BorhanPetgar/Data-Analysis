```python
import matplotlib.pyplot as plt
import numpy as np

pi = np.pi
e = np.e
```

```python
# plot point to point

xpoints = np.array([1, 10, 9, 20])
ypoints = np.array([0, 15, 30, 15])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.random.randint(1,10,20)
plt.plot(ypoints, 'o')
plt.show()
# The x-points in the example is [0, 1, 2, ..., 19].
```

```python
ypoints = np.random.randint(1,10,5)
plt.plot(ypoints, linestyle='dashed', linewidth='5',color='green', 
         marker='*', mfc='red', markeredgecolor='black', ms=20)
plt.show()
```

```python
# multiple plots
# 1. you can add more plt.plot(...)
# 2. use this format: plt(x1,y1 , x2,y2)

x1 = np.linspace(0,10,5)
y1 = np.random.randint(0,10,5)

x2 = np.linspace(0,10,4)
y2 = np.random.randint(0,10,4)

plt.plot(x1,y1, x2,y2)
plt.show()
```

```python
# Labels
x1 = np.array([10,15,20,25,30,40])
y1 = np.array([100,50,150,300,200,150])
font_properties = {'family':'serif','color':'darkred','size':15}

plt.plot(x1,y1, color='blue', linestyle='dashdot', marker='o', mec='green', mfc='green')
plt.xlabel('Year', fontdict=font_properties)
plt.ylabel('Energy consumed', fontdict=font_properties)
plt.title('Energy Consumption according to date', loc='center', fontdict=font_properties)
plt.show()
```

```python
# grid, legend
x = np.linspace(-3*pi,3*pi,1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color='blue', linestyle='solid', label='sin(x)')
plt.plot(x, y2, color='red', linestyle='solid', label='cos(x)')
plt.grid(color='grey', linestyle='dashed')
plt.xlabel('x', fontdict=font_properties)
plt.ylabel('y', fontdict=font_properties)
plt.title('Sin(x)', fontdict=font_properties)
plt.legend()
plt.show()
```

```python
# subplot

x1 = np.linspace(-5, 5, 1000)
y1 = (e) ** x1
x2 = np.linspace(0.01, 5, 1000)
y2 = np.log(x2)
x3 = np.linspace(-5, 5, 1000)
y3 = 1 / ((e ** (-x3)) + 1)
x4 = np.linspace(-0.25, 0.25, 1000)
y4 = 1 / x4

plt.subplot(2,2,1)
plt.plot(x1, y1, color='red')
plt.grid(linestyle='dashed', color='grey')
plt.subplot(2,2,2)
plt.plot(x2, y2, color='green')
plt.grid(linestyle='dashed', color='grey')
plt.subplot(2,2,3)
plt.plot(x3, y3, color='pink')
plt.grid(linestyle='dashed', color='grey')
plt.subplot(2,2,4)
plt.plot(x4, y4, color='blue')
plt.grid(linestyle='dashed', color='grey')
plt.show()
```

```python
# colorbar
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = y  # assign colors to y in order to get colors corresponding to the value

plt.scatter(x, y, c=y, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# bar chart
x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.random.randint(100, 500, 5)

plt.bar(x, y)
plt.show()
```
